# Explore

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fringe_all = pd.read_csv("fringe_all.csv")
venue_fringe_all = pd.read_csv('venue_fringe_all.csv')
disabled_fringe_all = pd.read_csv('disabled_fringe_all.csv')
#venue_fringe_all.head()
```

```python
chart_year = sns.countplot(x="year", data=fringe_all)
chart_year.set_title("Number of events held each year")
plt.show()
```

The number of fringe events held in three years has increased year by year.

```python
chart_venue1 = sns.countplot(x="has_booking_over_web", hue="year", data=venue_fringe_all)
chart_venue1.set_title("Whether booked online")
plt.show()
```

```python
chart_venue2 = sns.countplot(x="has_bar", hue="year", data=venue_fringe_all)
chart_venue2.set_title("Whether to provide a bar")
plt.show()
```

```python
chart_venue3 = sns.countplot(x="has_cafe", hue="year", data=venue_fringe_all)
chart_venue3.set_title("Whether to provide a cafe")
plt.show()
```

* People prefer to make reservations on the Internet, indicating that the Internet is the main source of people’s understanding of the event, and it can increase the publicity on the Internet, thereby increasing the popularity of the event.
* Most activities will choose to provide bar and cafe, indicating that this is also one of the main factors affecting the number of participants. The quality of the bar and cafe provided can be improved to increase the number of people participating in the event.

```python
chart_position = sns.relplot(data=fringe_all, x="longitude", y="latitude", col="year")
print("Location of the three-year event")
plt.show()
```

```python
chart_position2 = sns.jointplot(data=fringe_all, x="longitude", y="latitude", marker="+", s=100, marginal_kws=dict(bins=25))
plt.show()
```

* The three-year activities are concentrated in the location: near longitude -3.2, latitude 55.95. Explain that this is the place where people go most often, so holding an event here will get the most participants and spread.
* But at the same time, there are also activities held in other places because there are still people who can't go to the city center to participate in activities. And there will be people who prefer less crowded activities.

```python
chart_position3 = sns.jointplot(data=fringe_all, x="longitude", y="latitude")
chart_position3.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
chart_position3.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
plt.show()
```

This is the location of the event that contains the cluster. You can see the clusters of the main locations where the event is held. It also represents several major locations where people move.

```python
chart_country = sns.countplot(y="country", data=fringe_all)
chart_country.set_title("Activity country source")
plt.show()
```

Most of the activities are from the UK, and a few activities are from other countries.
